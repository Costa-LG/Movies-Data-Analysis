# Limpando Dados

## Limpando Texto

Parece haver um problema com a coluna `Star4`. Iremos aplicar a seguinte função, equivalente ao `CLEAN` do excel

```python
def clean(text):
    # check if the current value is null, if so return it
    if pd.isna(text):
        return text
    
    # remove non printable characters
    return re.sub(r'[\x00-\x1F]+', '', text)
```

```python
# apply the cleaning function to star4 column
df["Star4"] = df["Star4"].apply(clean)
```

## Conversão de Tipo

Como podemos ver no código a seguir, alguns atributos que deveriam ser numéricos são do texto.

```python
print(f"Released Year: {type(df["Released_Year"][0])}\n\
Runtime: {type(df["Runtime"][0])}\n\
IMDB Rating: {type(df["IMDB_Rating"][0])}\n\
Meta Score: {type(df["Meta_score"][0])}\n\
Number of Votes: {type(df["No_of_Votes"][0])}\n\
Gross Income: {type(df["Gross"][999])}")
```

Primeiro precisamos remover todos os carácteres não numéricos dos atributos. Para a coluna `runtime`, precisamos remover a string `' min'`, enquanto para o gross income, removeremos a vírgula `','`.

```python
# Here we define the necessary functions with examples of execution
def strip_min(text):
    return text.replace(' min','')
def strip_comma(text):
    # escape for the function if there are float values in gross attribute
    if float == type(text):
        return text
    return text.replace(',','')


runtime_example_text = '105 min'
runtime_example_text = strip_min(runtime_example_text)
print(runtime_example_text)

gross_example_text = '534,858,444'
gross_example_text = strip_comma(gross_example_text)
print(gross_example_text)
```

Aplicando estas funções às respectivas colunas, temos

```python
df["Runtime"] = df["Runtime"].apply(strip_min)
df["Gross"] = df["Gross"].apply(strip_comma)
```

```python
# Converting
conv_dict = {"Released_Year": 'Int64', "Runtime": 'Int64', "Gross": 'Int64'}
df = df.astype(conv_dict)
print(df.dtypes)
```

# Separando Dados

Agora que este conjunto de dados está limpo, quero organizar os dados, separando-os em tabelas diferentes, de acordo com o ER diagram do arquivo `DESIGN.md`.

```python
plots = pd.DataFrame(data = df, columns = ["id", "Overview"])
plots
```

```python
df.Director.describe()
```

```python
data_dict = {"Director": df.Director.unique()}
directors = pd.DataFrame(data = data_dict, columns = ["Director"])
directors.index += 1
directors.reset_index(inplace = True)
directors.rename(columns = {"index": "id"}, inplace = True)
directors
```

```python
actors.reset_index(inplace = True)
actors.rename(columns = {"index":"id"}, inplace = True)
actors.id += 1
actors
```

```python
acted = df.melt(id_vars = ["id"], value_vars = ["Star1", "Star2", "Star3", "Star4"], value_name = "actor")
acted = acted.merge(actors, on = "actor")
acted
```

```python
acted = pd.DataFrame(data = acted, columns = ["id_x", "id_y"])
acted.columns = ["movie_id", "actor_id"]
acted
```
